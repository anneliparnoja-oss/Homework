```python
text = """
Frankenstein is a novel about ambition, responsibility, and the limits of human knowledge.
Victor Frankenstein creates life but refuses responsibility for his creation.
The creature experiences rejection, loneliness, and anger.
The story explores themes of science, morality, isolation, and the consequences of unchecked ambition.
Fear, suffering, and revenge drive the narrative forward.
"""
```

```python
import urllib.request
import string
from collections import Counter

# 2. Basic cleaning
text = text.lower()
text = text.translate(str.maketrans("", "", string.punctuation))

# 3. Split into words
words = text.split()

# 4. Count word frequencies
word_counts = Counter(words)

# 5. Show basic stats
print("Total words:", len(words))
print("Unique words:", len(word_counts))
print("10 most common words:", word_counts.most_common(10))
```

```python
"""
Interpretation of the word cloud

The word cloud shows the words that appear most often in the text after removing common stopwords.
Words like responsibility, creation, ambition, and fear stand out, which suggests that the text focuses on moral themes rather than action.
Overall, the word cloud helps identify the main ideas of the text, but it does not show the full meaning or context."""
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

stopwords = STOPWORDS.union({
    "the","and","to","of","a","in","is","for","his","but"
})

wc = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=stopwords,
    max_words=50
).generate(" ".join(words))

plt.figure(figsize=(12, 5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
